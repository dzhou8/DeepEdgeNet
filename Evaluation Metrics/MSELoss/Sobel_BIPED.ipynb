```
import os
import imageio
import numpy as np
```

```
def mse(imageA, imageB):
	# the 'Mean Squared Error' between the two images is the
	# sum of the squared difference between the two images;
	# NOTE: the two images must have the same dimension
	err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
	err /= float(imageA.shape[0] * imageA.shape[1])
	
	# return the MSE, the lower the error, the more "similar"
	# the two images are
	return err
```

```
gdrive_path = '/content/drive/My Drive/CS 230/Project/'
labels_path = gdrive_path + "datasets/BIPED/edges/edge_maps/train/rgbr/real/"
output_path = gdrive_path + "Canny Detector/Canny-Edge BIPED/sobel_output/"

img_name = os.listdir(labels_path)
outputs = os.listdir(output_path)

totalMSE = 0
img_name.sort()
outputs.sort()
assert len(img_name) == len(outputs)

for i in range(len(img_name)):
  assert img_name[i].split('.')[0] == outputs[i].split('.')[0]
  label_img = imageio.imread(labels_path + img_name[i])
  output_img = imageio.imread(output_path + outputs[i])
  totalMSE += mse(label_img, output_img)
  print(img_name[i], mse(label_img, output_img))

averageMSE = totalMSE / len(img_name)
print("average MSE is " + str(averageMSE) + " over " + str(len(img_name)) + " images.")
```
