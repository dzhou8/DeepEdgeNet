```python
import os
import imageio
import numpy as np
```

```python
import sklearn.metrics

def flatten(nested_list):
    return np.asarray(nested_list).flatten()

def ODS(label, output, threshold):
	label_matrix = flatten(label.astype('float'))
	label_matrix = (label_matrix == 255)
	output_matrix = flatten(output.astype('float'))
	threshold_matrix = (output_matrix >= threshold)
	ODS_result = sklearn.metrics.f1_score(label_matrix, threshold_matrix)
	# print(ODS_result)
	
	return ODS_result
```

```python
import time
from statistics import mean

gdrive_path = '../../'
labels_path = gdrive_path + "datasets/BIPED/edges/edge_maps/test/rgbr/"
output_path = gdrive_path + "Model Results/BIPED2BIPED/vis/"
save_path = gdrive_path + "Evaluation Metrics/Optimal Dataset Scale (ODS)/DeepEdgeNet_BIPED.txt"

img_name = os.listdir(labels_path)
outputs = os.listdir(output_path)

img_name.sort()
outputs.sort()
assert len(img_name) == len(outputs)

if os.path.exists(save_path):
  save_file = open(save_path, 'r+')
else:
  save_file = open(save_path, 'w+')
lines = save_file.readlines()


ODS_results = []
for threshold in range(256):
  if threshold < len(lines):
    ods_average = float(lines[threshold].split(',')[1])
  else:
    for i in range(len(img_name)):
      thisThresholdResult = []
      assert img_name[i].split('.')[0] == outputs[i].split('.')[0]
      label_img = imageio.imread(labels_path + img_name[i])
      output_img = imageio.imread(output_path + outputs[i])
      ods = ODS(label_img, output_img, threshold)
      thisThresholdResult.append(ods)
    ods_average = mean(thisThresholdResult)
    save_file.write(str(threshold) + ',' + str(ods_average) + '\n')
    
  ODS_results.append(ods_average)

  t = time.localtime()
  current_time = time.strftime("%H:%M:%S", t)
  print(current_time, threshold, ods_average)

bestODS = max(ODS_results)
print("ODS is " + str(bestODS) + " over " + str(len(img_name)) + " images.")
```
